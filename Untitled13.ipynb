```python
import os
import json
import torch
import numpy as np
from PIL import Image
from transformers import AutoProcessor, AutoModelForZeroShotImageClassification
import faiss
import matplotlib.pyplot as plt

# Step 1: Load CLIP Model
processor = AutoProcessor.from_pretrained("openai/clip-vit-large-patch14")
model = AutoModelForZeroShotImageClassification.from_pretrained("openai/clip-vit-large-patch14").eval()

# Step 2: Load Metadata
metadata_path = "/content/drive/MyDrive/test_data.json/test_data.json"
images_folder = "/content/drive/MyDrive/test"
with open(metadata_path, "r") as f:
    data = [json.loads(line) for line in f.readlines()[:10]]  # Load first 100 entries

# Step 3: Initialize FAISS index
embedding_dim = 768 * 2  # Text (768) + Image (768) embedding size
index = faiss.IndexFlatL2(embedding_dim)  # L2 distance-based index
metadata_store = []  # To store metadata for each embedding

# Step 4: Generate and Store Embeddings
for entry in data:
    image_path = entry["image_path"]
    text = f"{entry['class_label']} {entry['brand']} {entry['product_title']}"
    metadata = {
        "image_path": image_path,
        "brand": entry["brand"],
        "product_title": entry["product_title"],
        "class_label": entry["class_label"]
    }

    # Generate text embedding
    text_inputs = processor(text=[text], return_tensors="pt", padding=True)
    with torch.no_grad():
        text_embedding = model.get_text_features(**text_inputs).cpu().numpy()

    # Generate image embedding
    image_full_path = os.path.join(images_folder, os.path.basename(image_path))
    image = Image.open(image_full_path).convert("RGB")
    image_inputs = processor(images=image, return_tensors="pt")
    with torch.no_grad():
        image_embedding = model.get_image_features(**image_inputs).cpu().numpy()

    # Combine text and image embeddings
    combined_embedding = np.hstack((text_embedding, image_embedding))

    # Add to FAISS index
    index.add(combined_embedding)
    metadata_store.append(metadata)

print("Embeddings stored in FAISS!")

# Step 5: Query System
def search_and_display(query_text, top_k=5):
    # Generate text embedding for the query
    text_inputs = processor(text=[query_text], return_tensors="pt", padding=True)
    with torch.no_grad():
        query_text_embedding = model.get_text_features(**text_inputs).cpu().numpy()

    # Pad the text embedding with zeros to match combined embedding dimensions
    query_embedding = np.hstack((query_text_embedding, np.zeros((1, 768))))

    # Perform FAISS search
    distances, indices = index.search(query_embedding, top_k)

    # Display results
    for idx in indices[0]:
        metadata = metadata_store[idx]
        image_path = metadata["image_path"]
        image_full_path = os.path.join(images_folder, os.path.basename(image_path))

        # Open and display the image
        image = Image.open(image_full_path)
        plt.imshow(image)
        plt.axis("off")
        plt.title(
            f"Brand: {metadata['brand']}\nClass: {metadata['class_label']}\nTitle: {metadata['product_title']}"
        )
        plt.show()

# Example Query
query = "saree with blouse peice"
search_and_display(query)
```
